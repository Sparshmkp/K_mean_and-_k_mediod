```python
#import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv("C:\\Users\HP\Downloads\cancer (1).csv")
df
```

```python
# Remove "id","diagnosis","Unnamed: 32" columns
df = df.drop(["id", "diagnosis","Unnamed: 32"], axis=1).values
```

```python
# Number of clusters
k = 2

# Helper function to calculate Euclidean distance between two points
def manhattan_distance(p1, p2):
    return np.sum(np.abs(p1 - p2))
```

```python
#Randomly assign starting positions to the cluster medoids
np.random.seed(0)
medoids_idx = np.random.choice(df.shape[0], k, replace=False)
medoids = df[medoids_idx]

#Initialize the assignments of data points to clusters.
labels=np.zeros(df.shape[0])
```

```python
# Maximum number of iterations
max_iters = 100
# Conduct K-Medoid clustering
for _ in range(max_iters):
    # Assign each data point to the closest medoid
    for i in range(df.shape[0]):
        distances = [manhattan_distance(df[i], medoid) for medoid in medoids]
        labels[i] = np.argmin(distances)

    # Revise medoids by choosing the data point within the same cluster that exhibits the least total distance to other points.
    new_medoids = np.copy(medoids)
    for i in range(k):
        cluster_points = df[labels == i]
        total_distances = np.array([np.sum([manhattan_distance(x, p) for p in cluster_points]) for x in cluster_points])
        new_medoids[i] = cluster_points[np.argmin(total_distances)]

    # verify if convergence has achieved
    if np.allclose(medoids, new_medoids):
        break

    medoids = new_medoids
```

```python
# Generate a visualization of K-Medoid clusters, distinguishing each cluster with distinct colors.
plt.figure(figsize=(8, 6))
for cluster_num in range(k):
    plt.scatter(df[labels == cluster_num, 0], df[labels == cluster_num, 1], label=f'Cluster {cluster_num + 1}')
plt.scatter(medoids[:, 0], medoids[:, 1], c='red', marker='X', s=100, label='Medoids')
plt.title('K-Medoid Clustering')
plt.xlabel('Radius Mean')
plt.ylabel('Texture Mean')
plt.legend()

# Count the number of points in each cluster
cluster_counts = [np.sum(labels == cluster_num) for cluster_num in range(k)]
print("Cluster Counts:")
for i, count in enumerate(cluster_counts):
    print(f'Cluster {i + 1}: {count} points')

# Show the plot
plt.show()
```
