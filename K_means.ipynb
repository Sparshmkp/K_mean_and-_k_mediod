```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('cancer (1).csv')
df
```

```python
# Remove "id","Unnamed: 32" and "diagnosis" columns
df = df.drop(["id", "diagnosis","Unnamed: 32"], axis=1).values
```

```python
# Number of clusters
k = 2

#Create a function that computes the Euclidean distance between two points.
def euclidean_distance(P1, P2):
    return np.sqrt(np.sum((P1 - P2) ** 2))
```

```python
#Randomly assign starting positions to the cluster centroids.
np.random.seed(0)
centroids = df[np.random.choice(df.shape[0], k, replace=False)]
```

```python
#Initialize the assignments of data points to clusters.
labels = np.zeros(df.shape[0])
```

```python
#The maximum allowed number of iterations
max_iters = 100

#Conduct K-Means clustering
for _ in range(max_iters):
    #Assign each data point to the closest centroid.
    for i in range(df.shape[0]):
        distances = [euclidean_distance(df[i], centroid) for centroid in centroids]
        labels[i] = np.argmin(distances)

    #Update the centroids by computing the mean of data points within each cluster.
    new_centroids = np.array([df[labels == i].mean(axis=0) for i in range(k)])

    #Verify if convergence has been achieved.
    if np.allclose(centroids, new_centroids):
        break

    centroids = new_centroids
```

```python
#Generate a visualization of K-Means clusters, using distinct colors for each cluster.
plt.figure(figsize=(8, 6))
for cluster_num in range(k):
    plt.scatter(df[labels == cluster_num, 0], df[labels == cluster_num, 1], label=f'Cluster {cluster_num + 1}')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=100, label='Centroids')
plt.title('K-Means Clustering')
plt.xlabel('Radius Mean')
plt.ylabel('Texture Mean')
plt.legend()
```

```python
# Count the number of points in each cluster
cluster_counts = [np.sum(labels == cluster_num) for cluster_num in range(k)]
print("Cluster Counts:")
for i, count in enumerate(cluster_counts):
    print(f'Cluster {i + 1}: {count} points')

# Show the plot
plt.show()
```
